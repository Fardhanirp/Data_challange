### data['Value'], data['Wage'], and data['Release Clause']

```python
#Removing the currency symbol and convert M to millon and K to thousands in numbers
def convert_currency(amount):
    amount = amount.replace('€','') #removing currency symbol
    if 'M' in amount:
        amount = float(amount.replace('M','')) * 1000000 # convert to million
    elif 'K' in amount:
        amount = float(amount.replace('K','')) * 1000 #convert to thousands
    else:
        amount = float(amount)
    return amount

data['Value'] = data['Value'].apply(convert_currency)
data['Wage'] = data['Wage'].apply(convert_currency)
data['Release Clause'] = data['Release Clause'].apply(convert_currency)
```

```python
data[['Value','Wage','Release Clause']]
```

### data['WF'], data['SM'], and data['IR']

```python
#removing the star symbol in w/f,sm.ir
data['W/F'] = data['W/F'].str.replace('★','')
data['SM'] = data['SM'].str.replace('★','')
data['IR'] = data['IR'].str.replace('★','')
```

### data['Loan Date End'] and data['Contract']

Loan Date End is the date when the player's loan contract ends. Some values are null (NaN) since the contract has expired as we see in Contract column, but some remain on loan and free.

```python
data['Loan Date End'].unique()
```

```python
data[data['Contract']=='Free'][['Contract','Loan Date End']]
```

```python
data[['month','Contract Expiry Year']] = data['Loan Date End'].str.split(',', expand = True).apply(lambda x: x.str.strip())
data.drop(['month','Loan Date End'] , axis =1, inplace = True)
data.head()
```

```python
data['Contract Expiry Year'].unique()
```

Fill the NaN values with year by extract 4 characters (year) from Contract column 

```python
end_year= data['Contract'].apply(lambda x : x[-4:])

data['Contract Expiry Year'] = data['Contract Expiry Year'].fillna(end_year)
data['Contract Expiry Year'].replace('Free','0',inplace=True)
data['Contract Expiry Year'].unique()
```

```python
data['Hits'].unique()
```

```python
data['Hits'] = data['Hits'].fillna(0)
data['Hits'] = data['Hits'].replace({'k':'000','K':'000'}, regex = True).astype(float)
#Updated data
data[['Hits']]
```
